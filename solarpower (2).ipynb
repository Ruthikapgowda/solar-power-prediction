```python
import matplotlib.pyplot as plt
```

```python
# plot distribution of power
plt.figure(figsize=(10,6))
sns.histplot(df['generated_power_kw'], bins=30, kde=True)
plt.title('Distribution of Generated Power (kW)')
plt.xlabel('Generated Power (kW)')
plt.ylabel('Frequency')
plt.show()
```

```python
# EDA
# histograms for remaining columns
df[df.columns[:9]].hist(bins=30, figsize=(15, 10))
plt.show()
```

```python
df[df.columns[18:20]].hist(bins=30, figsize=(10, 5))
plt.show()
```

```python
# Bivariate Analysis
# Scatter plot with target feature
plt.figure(figsize=(15, 30))
for i, column in enumerate(df.columns):
    plt.subplot(7, 3, i+1)
    plt.scatter(df[column], df['generated_power_kw'])
    plt.title(f'{column} vs Generated Power (kW)')
    plt.xlabel(column)
    plt.ylabel('Generated Power (kW)')
plt.tight_layout()
plt.show()
```

```python
df.corr()
```

```python
plt.figure(figsize=(15, 10))
sns.heatmap(df.corr(), cmap='coolwarm', annot=True, fmt='.2f')
plt.title('Feature Correlation Heatmap')
plt.show()
```

```python
# Outlier
plt.figure(figsize=(15, 30))
for i, column in enumerate(df.columns):
    plt.subplot(7, 3, i+1)
    sns.boxplot(df[column])
plt.show()
```

```python
plt.figure(figsize=(15, 30))
columns_to_plot=df.columns[:-1]
for i, column in enumerate(columns_to_plot):
    plt.subplot(7, 3, i+1)
    plt.scatter(df[column], df['generated_power_kw'])
    plt.title(f'{column} vs Generated Power (kW)')
    plt.xlabel(column)
    plt.ylabel('Generated Power (kW)')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 30))
for i, column in enumerate(df.columns,):
    plt.subplot(7, 3, i+1)
    sns.boxplot(df[column], color='aqua')
    plt.grid(axis='y')
    plt.title(f'Boxplot of {column}')
plt.tight_layout()
plt.show()
```
